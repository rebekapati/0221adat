```sql
SELECT
    IIF(v.USERNEV IS NULL, 'Végösszeg', v.USERNEV),
    COUNT(DISTINCT szh.SZALLAS_ID) AS 'Külön szálláshelyek száma'
FROM Vendeg v LEFT JOIN Foglalas f ON v.USERNEV = f.UGYFEL_FK
              LEFT JOIN Szoba sz ON f.SZOBA_FK = sz.SZOBA_ID
              LEFT JOIN szallashely szh ON sz.SZALLAS_FK = szh.SZALLAS_ID
GROUP BY ROLLUP(v.USERNEV)
```

```sql
SELECT *,
    YEAR(GETDATE())-YEAR(SZUL_DAT) AS 'Vendég életkora',
    AVG(YEAR(GETDATE())-YEAR(SZUL_DAT))
        OVER(ORDER BY SZUL_DAT ROWS BETWEEN 2 PRECEDING AND 2 FOLLOWING) AS '4-es blokk átlagolt életkor'
FROM Vendeg
```

```sql
SELECT DISTINCT v.USERNEV, v.NEV
FROM Vendeg v JOIN Foglalas f ON v.USERNEV=f.UGYFEL_FK
WHERE NOT EXISTS
    (SELECT *
    FROM Foglalas f2
    WHERE f2.UGYFEL_FK=v.USERNEV AND f2.FELNOTT_SZAM + f2.GYERMEK_SZAM <> 2)
ORDER BY v.USERNEV
```
